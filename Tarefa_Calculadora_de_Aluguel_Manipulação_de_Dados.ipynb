## Tarefa: Calculadora de Aluguel - Manipulação de Dados



```python
# Bibliotecas
import pandas as pd
```

```python
# Carregando arquivo CSV
df = pd.read_csv('/content/dados_mercado_livre (1).csv')
```

```python
# As cincos primeiras linhas
df.head()
```

```python
# Ajustando as colunas e suas informações
df['aluguel'] = df.aluguel.str.split('/Mês').str[0]
df['aluguel'] = df['aluguel'].map(lambda x: x.lstrip('R$ ').rstrip(''))
df['area'] = df.area.str.split('m²').str[0]
df['quartos'] = df.quartos.str.split('Quartos' and 'Quarto').str[0]
df['quartos'] = df.quartos.str.split('Quartos' and 'Quarto').str[0]
df.head()
```

```python
# Transformando os dados que eu preciso em inteiros
df['aluguel'] = df['aluguel'].astype(int)
df['area'] = df['area'].astype(int)
df.info()
```

```python
df.to_csv('dados_tratados_ml.csv', index=False)
pd.read_csv('dados_tratados_ml.csv')
```
