## Tarefa: Calculadora de Aluguel - Manipulação de Dados



```python
# Bibliotecas
import pandas as pd
```

```python
# Carregando arquivo CSV
df = pd.read_csv('/content/dados coletados (Mercado Livre).csv')
```

```python
# As cincos primeiras linhas
df.head()
```

```python
# Ajustndo as minhas colunas e meus dados
df [['area', 'quarto']] = df['areas_quartos'].str.split('construídos', expand=True)
df['precos'] = df['precos'].replace('\.', '', regex=True)
df['area'] = df.area.str.split('m²').str[0]
df['area'] = df.area.str.split(',').str[0]
df['quarto'] = df.quarto.str.split('quartos' and 'quarto').str[0]
df['quarto'] = df.quarto.fillna(0)
df= df.drop(columns=['areas_quartos'])
df = df.drop(columns=['enderecos'])
df.head()
```

```python
# Transformando os dados em floats
df['quarto'] = df['quarto'].astype(float)
df['area'] = df['area'].astype(float)
df.info() 
```

```python
# Verificando a quantidade dos dados
df.shape
```

```python
df.to_csv('dados_tratados_ml.csv', index=False)
pd.read_csv('dados_tratados_ml.csv')
```
